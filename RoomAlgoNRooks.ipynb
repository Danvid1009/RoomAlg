```python
def n_queens(n):
    def is_valid(queens, new_queen):
        for row, col in enumerate(queens):
            if col == new_queen:
                return False
        return True

    def solve(queens):
        if len(queens) == n:
            board = [[0] * n for _ in range(n)]
            for i, j in enumerate(queens):
                board[i][j] = 1
            results.append(board)
            return
        for i in range(n):
            if is_valid(queens, i):
                solve(queens + [i])

    results = []
    solve([])

    return results

n = 7
solutions = n_queens(n)
```
